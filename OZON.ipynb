```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy.random as rand
plt.style.use('ggplot')
```

```python
dns = sorted(np.random.randint(6000, 120000, 700))
```

```python
ozon = sorted(np.random.randint(6000, 120000, 700))
```

```python
df = pd.DataFrame({'ozon': ozon, 'dns': dns})
```

```python
df
```

```python
df['diff'] = df.ozon / df.dns
```

```python
df
```

```python
plt.rcParams['figure.figsize'] = (30, 20)
fig, ax = plt.subplots()
sns.lineplot(x=df.index, y='diff', data=df.sort_values('diff', ascending=False), marker='o', linewidth=3)
y = [1, 1, 1]
x = [0, 350, 700]
plt.plot(x, y, linewidth=7, color=(0.2, 0.4, 0.6, 1.0))
plt.title('', fontsize=30, pad=40)
plt.xlabel('Модель смартфона', fontsize=30, labelpad=30)
plt.ylabel('Отношение цен', fontsize=30, labelpad=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
#ax.text(110, 1.3, '← Превышение цены на 30%', fontsize=30)
ax.text(110, 1.4, '← Превышение цены на 40%', fontsize=30)
#ax.text(110, 1.2, '← Превышение цены на 20%', fontsize=30)
ax.text(130, 1.1, '← Превышение цены на 10%', fontsize=30)
ax.text(400, 1.005, 'Рынок', fontsize=40)

box = {'facecolor':(0.2, 0.4, 0.6, 1.0),    #  цвет области
       'edgecolor': 'red',     #  цвет крайней линии
       'boxstyle': 'rarrow',    #  стиль области
         'pad': 0.9}
ax.text(215, 0.95, 'Ниже рынка',
        bbox = box,
        color = 'white',
        fontsize = 30)
box2 = {'facecolor':(0.2, 0.4, 0.6, 1.0),    #  цвет области
       'edgecolor': 'red',     #  цвет крайней линии
       'boxstyle': 'larrow',    #  стиль области
         'pad': 0.9
      }
ax.text(105, 1.25, 'Выше рынка',
        bbox = box2,
        color = 'white',
        fontsize = 30)
sns.set_style("whitegrid")
paper_rc = {'lines.linewidth': 1, 'lines.markersize': 5}                  
sns.set_context("paper", rc = paper_rc)

sns.despine()
```
